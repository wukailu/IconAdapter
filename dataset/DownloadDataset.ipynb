```python
import requests
import wget
import json
import zipfile
import os
import shutil
URL_path = "theme_list.txt"
ban_list = ['icon_folder.png',
            'icon_mask.png',
            'icon_pattern.png']
```

```python
cnt = 0
totalcnt = 0
for line in open(URL_path, 'r'):
    info = requests.get(line.replace("http://zhuti.xiaomi.com/detail/", "http://thm.market.xiaomi.com/thm/download/v2/"))
    info = json.loads(info.text)
    if info['apiCode'] != 0:
        print("Getting Url Error!!!")
    else:
        url = info['apiData']['downloadUrl']
        wget.download(url, out = "./tmp.zip")
        file = zipfile.ZipFile("./tmp.zip")
        file.extract("icons", ".")
        file = zipfile.ZipFile("./icons")
        file.extractall("./tmp1")
        file.close()
        base = "./tmp1/res/drawable-xxhdpi/"
        os.makedirs("./dataset/" + str(cnt))
        for pngs in os.listdir(base):
            if pngs.endswith(".png") and not pngs in ban_list:
                shutil.copy(base+pngs, "./dataset/" + str(cnt) + "/" + pngs)
                totalcnt += 1
        
        # clean up
        os.remove("./tmp.zip")
        os.remove("./icons")
        shutil.rmtree("./tmp1")
        
        cnt += 1
    
print("totally %d themes and %d icons are downloaded"%(cnt, totalcnt))
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm
import warnings
from collections import Counter
```

```python
with warnings.catch_warnings():
    warnings.filterwarnings('error')
    shapes = []
    for dirs in tqdm(os.listdir("./dataset")):
        for pngs in os.listdir("./dataset/"+dirs):
            try:
                path = "./dataset/"+dirs+"/"+pngs
                if path.endswith(".png"):
                    img = cv2.imread(path, flags = cv2.IMREAD_UNCHANGED)
                else:
                    print("not png", path)
                shapes.append(img.shape)
            except Warning:
                print(path, "get a Warning")
            except Exception:
                print(path)
                os.remove(path)
    print(Counter(shapes))
```
